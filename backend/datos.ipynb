```python
import pandas as pd
import numpy as np
```

```python
# Creo df desde CSV y hago copia para no alterar original 

df_datos_full_original = pd.read_csv('data/hackathon_startup_week.csv') 

df_datos_full = df_datos_full_original.copy()
```

```python
print(df_datos_full.dtypes)
```

```python
# Dropeo trans_num por irrelevante y para reducir tamaño de CSV

df_datos_full.drop('trans_num', axis=1, inplace=True)
```

```python
df_datos_full.info()
```

```python
df_datos_full.head(10)
```

```python
# Incluyo día de la semana a las fechas del df

unix_time_datetime = pd.to_datetime(df_datos_full['unix_time'], unit='s')
df_datos_full['weekday'] = unix_time_datetime.dt.day_name()
```

```python
# Separo los datos de la fecha en "Año", "Mes", "Día" y "Hora" y borro columna original 

df_datos_full['trans_date_trans_time'] = pd.to_datetime(df_datos_full['trans_date_trans_time'])
df_datos_full['year'] = df_datos_full['trans_date_trans_time'].dt.year
df_datos_full['month'] = df_datos_full['trans_date_trans_time'].dt.month
df_datos_full['day'] = df_datos_full['trans_date_trans_time'].dt.day
df_datos_full['time'] = df_datos_full['trans_date_trans_time'].dt.time
df_datos_full.drop(columns=['trans_date_trans_time'], inplace=True)
```

```python
df_datos_full.to_csv('data/datos_full.csv', index=False)
```

```python
# Transacciones por día de la semana y categoría

pd.set_option('display.max_rows', None)
df_sorted = df_datos_full.sort_values('category')
dia_cate = df_sorted.groupby('weekday')['category'].value_counts()
dia_cate
```

```python
df_dia_cate = dia_cate.reset_index().rename(columns={'weekday': 'weekday', 'category': 'category', 0: 'count'})
df_dia_cate_pivot = df_dia_cate.pivot(index='category', columns='weekday', values='count')
df_dia_cate_pivot = df_dia_cate_pivot.reindex(df_dia_cate_pivot.sum().sort_values(ascending=False).index, axis=1)
df_dia_cate_pivot.loc['Total'] = df_dia_cate_pivot.sum()
df_dia_cate_pivot['Total'] = df_dia_cate_pivot.sum(axis=1)
df_dia_cate_pivot = df_dia_cate_pivot.sort_index()
df_dia_cate_pivot
```

```python
pd.set_option('display.max_rows', None)
df_sorted = df_datos_full.sort_values('merchant')
cat_merch = df_sorted.groupby('category')['merchant'].value_counts()
cat_merch
```

```python
df_cat_merch = cat_merch.reset_index().rename(columns={'category': 'category', 'merchant': 'merchant', 0: 'count'})
df_cat_merch_pivot = df_cat_merch.pivot(index='category', columns='merchant', values='count')
df_cat_merch_pivot = df_cat_merch_pivot.reindex(df_cat_merch_pivot.sum().sort_values(ascending=False).index, axis=1)
df_cat_merch_pivot.loc['Total'] = df_cat_merch_pivot.sum()
df_cat_merch_pivot['Total'] = df_cat_merch_pivot.sum(axis=1)
df_cat_merch_pivot = df_cat_merch_pivot.sort_index()
df_cat_merch_pivot
```

```python
# Transacciones por día de la mes y categoría

pd.set_option('display.max_rows', None)
df_sorted = df_datos_full.sort_values('category')
mes_cate = df_sorted.groupby('mes')['category'].value_counts()
mes_cate
```

```python
# Se buscan transacciones recurrentes

def identify_recurrent_purchases(df, criteria1='merchant', criteria2='amt', frequency_threshold=1):
    grouped = df.groupby([criteria1, criteria2])
    
    # Create a DataFrame that includes the count of recurrent purchases for each group
    recurrent_purchases_counts = grouped.size().reset_index(name='count')
    
    # Filter the groups that have count greater or equal to frequency_threshold
    recurrent_purchases_counts = recurrent_purchases_counts[recurrent_purchases_counts['count'] >= frequency_threshold]
    
    return recurrent_purchases_counts

recurrent_purchases_counts = identify_recurrent_purchases(df_datos_full, criteria1='merchant', criteria2='amt', frequency_threshold=3)
recurrent_purchases_counts.tail()
```
