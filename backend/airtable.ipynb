```python
# Función para crear un DF a partir de un archivo CSV guardado en el directorio

def csv_a_df(archivo):
    data_frame = pd.read_csv(f"data\{archivo}.csv")
    
    # Reemplaza los valores nan por None en caso de haberlos
    data_frame = data_frame.replace({np.nan : None})
    
    return data_frame 
   
    
csv_a_df("hackathon_ordenado") 

# Guarda en variable df_datos el df de la lista de datos
df_datos = csv_a_df("hackathon_ordenado") 

df_datos
```

```python
# FUNCION PREPARACION DE DATOS Y CREACIÓN DE REGISTROS EN AIRTABLE A PARTIR DE DF 


def creacion_registros(df):
    
  
    i = 0 # Inicializa el contador de filas del df original (df_datos)
    j = 1 # Inicializa el contador de paginas de 10 filas cada una (df_paginas_datos)

# Crea un dict para almacenar dict de 10 filas cada uno hasta recorrer todas las filas del df original (df_datos)
    paginas_datos = {}
    

    while i < df.shape[0]:   # iniciliza un bucle while, ejecutable mientras i sea menor a la cantidad de filas 
        datos = {"records"  : [],
                   "typecast" : True}
    
        filas = df.iloc[i:i+10, : ]

        for _, fila in filas.iterrows():
            fila_dict = {"fields" : fila.to_dict()}  
            datos["records"].append(fila_dict)

            paginas_datos[f"datos_{j}"] = datos
        
        i += 10 # Incrementa el contador de filas del df original (df_datos) en 10 filas
        j += 1 # Incrementa el contador de paginas del df_paginas_datos en 1 pagina, para pasar a la siguiente pagina de 10 filas

   
    print("-"*120)

    for key, pagina in paginas_datos.items():  # Iterea sobre el diccionario de paginas_datos, posteando cada pagina en la API de Airtable mediante un bucle for
        response = requests.post(url = url_endpoint, json = pagina, headers = headers) # Postea los registros a la tabla de AirTable
 
          
    print(f"response: {response.status_code}")

    print(f"endpoint: {response.url}")
    
    total_registros = sum(len(pagina["records"]) for pagina in paginas_datos.values()) # Suma el total de registros en cada pagina para darme el número total de registros cargados
    
    print(f"Total de registros cargados: {total_registros}")

    print("-"*120)
        
    pprint(paginas_datos)
    
    sleep(0.5)
           
creacion_registros(df_datos)
```
